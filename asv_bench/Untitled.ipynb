```python
import pandas as pd
import numpy as np
import timeit

# Mixed dtypes force the split path where np.ndim is called
df = pd.DataFrame(
    {
        "a": np.zeros(100_000),
        "b": np.zeros(100_000, dtype=int),
        "c": np.zeros(100_000),
    }
)

val_list = np.random.randn(100_000, 2).tolist()
val_arr = np.array(val_list)

t_list = timeit.timeit(
    lambda: df.loc.__setitem__((slice(None), ["a", "c"]), val_list), number=100
)
t_arr = timeit.timeit(
    lambda: df.loc.__setitem__((slice(None), ["a", "c"]), val_arr), number=100
)

print(f"list-of-lists: {t_list:.3f}s")
print(f"ndarray:       {t_arr:.3f}s")
print(f"Ratio: {t_list / t_arr:.1f}x")
```
